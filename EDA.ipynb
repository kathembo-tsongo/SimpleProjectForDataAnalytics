### How to manipulate floats or numbers after the comma. You can for example want to have something like x.0 or x.00 or x.000. 
#### You can tell Pandas to select only columns that are already "numbers" (integers or other floats) and convert them all to float64. This completely avoids your text columns.

```python
numeric_cols = dataframe.select_dtypes(include=['number']).columns
dataframe[numeric_cols] = dataframe[numeric_cols].astype(float)
pd.set_option('display.float_format', lambda x: '%.2f' % x)
dataframe
```

### Do some visualization with heatmap
#### What is a heatmap?

A heatmap is a data visualization technique that uses color to represent the magnitude of values in a dataset, revealing patterns, trends, or correlations at a glance, often as a grid of colored squares or a colored overlay on a map, showing high-intensity areas in warmer colors (like red/yellow) and lower intensity in cooler colors (like blue/green)

```python
sns.heatmap(dataframe.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.rcParams['figure.figsize'] = (10, 6)
plt.title('Correlation Heatmap')
plt.show()
```

#### 1. df.corr(numeric_only=True): Calculates the math (correlation coefficients) for your numbers.
#### 2. annot=True: Labels the cells with those calculated numbers.
#### 3. cmap='coolwarm': Paints the high numbers red and the low numbers blue.

**Cool** refers to Blue (associated with cold temperatures).

**Warm** refers to Red (associated with heat).

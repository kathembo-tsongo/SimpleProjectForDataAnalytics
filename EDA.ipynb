### How to manipulate floats or numbers after the comma. You can for example want to have something like x.0 or x.00 or x.000.

### Do some visualization with heatmap
#### What is a heatmap?

A heatmap is a data visualization technique that uses color to represent the magnitude of values in a dataset, revealing patterns, trends, or correlations at a glance, often as a grid of colored squares or a colored overlay on a map, showing high-intensity areas in warmer colors (like red/yellow) and lower intensity in cooler colors (like blue/green)

```python
sns.heatmap(dataframe.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.rcParams['figure.figsize'] = (10, 6)
plt.title('Correlation Heatmap')
plt.show()
```

#### 1. df.corr(numeric_only=True): Calculates the math (correlation coefficients) for your numbers.
#### 2. annot=True: Labels the cells with those calculated numbers.
#### 3. cmap='coolwarm': Paints the high numbers red and the low numbers blue.

**Cool** refers to Blue (associated with cold temperatures).

**Warm** refers to Red (associated with heat).

### Look at data more closely
#### Let's find the continents with largest population

```python
dataframe.groupby('Continent').mean(numeric_only=True).sort_values(by='2022 Population', ascending=False)
```

### Now Try to visualize this result in a plot

```python
dataframe2 = dataframe.groupby('Continent').mean(numeric_only=True).sort_values(by='2022 Population', ascending=False)
dataframe2
```

```python
dataframe2.plot()
```

#### Normally we wanted to visualize not the Population but the continents but in the above plot we are visualizing population. We then need to transport the data by making raws to be columns and columns to be raws in out dataset, because in general logic columns are always the ones visualized by default

```python
dataframe3 = dataframe2.transpose()
dataframe3
```

```python
dataframe3.plot()
```

#### But this does not look good because the plot is visualizing everything. What we need to look at is only the data about the 2022 Population. So let us specify the targetted columns which in our case are Population for different years.

```python
dataframe2.columns
```

```python

dataframe4 = dataframe.groupby('Continent')[['2022 Population', '2020 Population', '2015 Population',
       '2010 Population', '2000 Population', '1990 Population',
       '1980 Population', '1970 Population']].mean(numeric_only=True).sort_values(by='2022 Population', ascending=False)
dataframe4
```

#### There is also a short way of doing this just by calling columns from the dataframe.

```python
dataframe4 = dataframe.groupby('Continent')[dataframe.columns[5:13]].mean(numeric_only=True).sort_values(by='2022 Population', ascending=False)
dataframe4
```

#### Now let's transpose again

```python
dataframe5 = dataframe4.transpose()
dataframe5
```

```python
dataframe5.plot()
plt.title('Average Population by Continent Over Years')
plt.xlabel('Year')
plt.ylabel('Average Population')
plt.xticks(rotation=45)
plt.show()
```

#### As compared to the previous one, this graph looks good now and can be easily interpreted. The only thing is that this is not the best way of doing it for years follow the backward order. Let's the come backt to the long form where we can manually arrange the data before visualization.

```python
dataframe6 =  dataframe.groupby('Continent')[['1970 Population', '1980 Population', '1990 Population',
       '2000 Population', '2010 Population', '2015 Population',
       '2020 Population', '2022 Population']].mean(numeric_only=True).sort_values(by='2022 Population', ascending=False)
dataframe6
```

#### Then transpose the order in the dataframe

```python
dataframe7 = dataframe6.transpose()
```

```python
dataframe7.plot()
plt.title('Average Population by Continent Over Years')
plt.xlabel('Year')
plt.ylabel('Average Population')
plt.xticks(rotation=45)
plt.show()
```

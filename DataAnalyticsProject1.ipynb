## SIMPLE PROJECT IN DATA ANALYTICS

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [24, 27, 22, 32, 29],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix'],
    'Salary': [70000, 80000, 60000, 90000, 75000]
}

df = pd.DataFrame(data)
df
```

```python
df.head()
```

```python
df.describe()
```

```python
df.groupby('Salary')
```

### Understand what Seaborn does

```python
sns.histplot(data=df, x='Age', y='Salary', hue='City')
plt.show()
```

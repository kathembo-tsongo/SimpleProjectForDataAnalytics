c. **A program that solves and shows the summation of 64 + 32**

```python
n1 = 64
n2 = 32
summation = n1 + n2

# call the variable summation
print("The summation of:", n1, "&", n2, "is", summation)
```

```python
# Another way of doing the same summation is by using the f-string or formatted string. The role of this is that we don't worry about commas
n1 = 64
n2 = 32
summation = n1 + n2
print(f"The summation of {n1} & {n2} is {summation}")
```

e. **A program that asks the user for a number of days. The program then
prints out the number of seconds in the amount of days given.**

```python
# Ask the user to input the number of days
days = float(input("Enter the number of days: "))

# Calculation: 1 day = 24 hours,
# 1 hour = 60 minutes,
# 1 minute = 60 seconds
# ==> 24 * 60 * 60 = 86,400 seconds per day
seconds_per_day = 86400
total_seconds = days * seconds_per_day

# Display the result using an f-string
print(f"There are {total_seconds:,} seconds in {days} days.")
```

f. **A program that asks a user to input the radius then the program
calculates the volume of a sphere (the formula for the volume is (4/3)πr3).
We shall be using the exponential operator in python to compute (r3)**

```python
import math

# Ask the user for the radius and convert it to a float
radius = float(input("Enter the radius of the sphere: "))

# Calculate the volume using the formula: (4/3) * pi * r^3
volume = (4 / 3) * math.pi * (radius ** 3)

# Display the result rounded to 2 decimal places
print(f"The volume of the sphere with radius {radius} is {volume:.2f}")
```
